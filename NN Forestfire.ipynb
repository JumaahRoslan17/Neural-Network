```python
import pandas as pd
forest = pd.read_csv('forestfires (2).csv')
```

```python
import numpy as np
```

```python
forest.info()
```

```python
forest.describe()
```

```python
forest.drop(["month","day"], axis = 1, inplace = True)
```

```python
def norm_func(i):
  x = (i-i.mean())/(i.std())
  return (x)
```

```python
df = norm_func(forest.iloc[:,0:8])
df.describe()
```

```python
model_df = pd.concat([forest.iloc[:,[8]],df,forest.iloc[:,9:28]], axis=1)
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
np.random.seed(10)
```

```python
X=model_df.iloc[:,1:]
Y=model_df.iloc[:,0]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size=0.2, random_state=457)
```

```python
import sklearn.metrics as skl_mtc
from tensorflow import keras
import matplotlib.pyplot as plt
```

```python
model = keras.models.Sequential()
model.add(keras.layers.Dense(5000, activation='relu', input_dim=27))
model.add(keras.layers.BatchNormalization())
model.add(keras.layers.Dense(500, activation='relu'))
model.add(keras.layers.BatchNormalization())
model.add(keras.layers.Dense(50, activation='relu'))
model.add(keras.layers.BatchNormalization())
model.add(keras.layers.Dense(1, kernel_initializer='uniform'))
model.compile(loss=keras.losses.MeanSquaredError(),optimizer=keras.optimizers.Nadam(learning_rate=0.0005,beta_1=0.8, beta_2=0.999),metrics=["mse"])
```

```python
early_stopping = keras.callbacks.EarlyStopping(monitor='val_loss',verbose=1,patience=20,mode='auto',restore_best_weights=True)
```

```python
reduce_lr = keras.callbacks.ReduceLROnPlateau(monitor='val_loss',factor=0.2,patience=10,verbose=1,mode='auto',min_delta=0.0005,cooldown=0,min_lr=1e-6)
```

```python
model.fit(x=x_train,y=y_train,batch_size=15,epochs=100)
```

```python
eval_score_test = model.evaluate(x_test,y_test,verbose=1)
```

```python
eval_score_train = model.evaluate(x_train,y_train,verbose=1)
```

```python
predict_y = model.predict(x_test)
```

```python
result = skl_mtc.r2_score(y_test, predict_y)
print(f'R2-score in test set: {np.round(result, 4)}')
```

```python
pred_df = pd.DataFrame(predict_y, columns =['predict_y'])
pred_y = pred_df.iloc[:,0]
```

```python
test_resid = pred_y - y_test
```

```python
test_rmse = np.sqrt(np.mean(test_resid * test_resid))
test_rmse
```

```python
history = model.fit(x_train, y_train, epochs=10, batch_size=50, verbose=1, validation_split=0.2)
```
